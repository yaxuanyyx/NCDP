```python
df1 = pd.read_csv('AWR370W_UserCourseTranscripts_2023-06-13-20-08-05.csv', low_memory=False)
df2 = pd.read_csv('AWR390W_UserCourseTranscripts_2023-06-13-20-05-52.csv', low_memory=False)
df3 = pd.read_csv('AWR391W_UserCourseTranscripts_2023-06-13-20-07-26.csv', low_memory=False)
```

```python
# filter for Stu17
df1 = df1[df1['Interaction Description'].isin(['Stu17 Please provide any other comments or suggestions you have for improving this course'])]
df2 = df2[df2['Interaction Description'].isin(['Stu17 Please provide any other comments or suggestions you have for improving this course'])]
df3 = df3[df3['Interaction Description'].isin(['Stu17 Please provide any other comments or suggestions you have for improving this course'])]
```

```python
# Delete rows with blank values in 'Learner Response' column
df1 = df1.dropna(subset=['Learner Response'], how='all')

# Create a list of values to remove
values_to_remove = ['!', '-', '.', 'c', 'N/A', 'na', 'none', 'N/a', 'NILL', 'no', 'No comments at all', 'no comments.', 'None',
                   'no thanks', 'None needed. ', 'x', 'No']

# Filter the DataFrame and keep rows where 'Learner Response' is not in the values to remove
df1 = df1[~df1['Learner Response'].isin(values_to_remove)]
```

```python
# define the sentiment label
sentiment_labels = {
    'positive': ['valuable', 'great', 'excellent', 'good', 'informative', 'helpful', 'perfect'],
    'negative': ['bad', 'poor', 'terrible'],
    'neutral': ['okay', 'ok', 'average', 'neutral', 'thank', 'enjoy', 'nothing'],
    'suggestion': ['suggest', 'improve', 'add', 'less', 'more', 'issue']
}

# assign sentiment label based on text
def assign_sentiment_label(text):
    # convert text to lowercase for case-insensitive matching
    text = text.lower()
    
    # positive sentiment keywords
    for key in sentiment_labels['positive']:
        if key in text:
            return 'positive'
    
    # negative sentiment keywords
    for key in sentiment_labels['negative']:
        if key in text:
            return 'negative'
    # suggestion sentiment keywords
    for key in sentiment_labels['neutral']:
        if key in text:
            return 'neutral'
    
    # If no positive or negative keywords found, assign neutral label
    return 'suggestion'
```

```python
# Iterate over each row in the DataFrame
for index, row in df1.iterrows():
    text = row['Learner Response']
    label = assign_sentiment_label(text)
    df1.at[index, 'Sentiment'] = label
```
