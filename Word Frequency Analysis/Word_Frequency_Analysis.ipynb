```python
import pandas as pd
import numpy as np

import nltk

from nltk.corpus import stopwords
from collections import Counter
from nltk.tokenize import word_tokenize
import re
```

```python
df = pd.read_csv('PER409 - EVALUATION_September 21, 2023_15.10.csv', low_memory=False)
```

```python
# Keep Q8, Q9, Q10 and drop all the other columns
columns_to_keep = ['Q10']
df = df[columns_to_keep]
```

```python
columns_to_check = ['Q10']

# Remove rows with blank values in these columns
df = df.dropna(subset=columns_to_check)
```

```python
# Delete row 0 and 1
index_to_delete = 1

df = df.drop(index_to_delete)
```

```python
index_to_delete_0 = 0

df = df.drop(index_to_delete_0)
```

```python
# Define the list of columns to analyze
columns_to_analyze = ['Q10']

# Tokenize the text and remove stopwords
stop_words = set(stopwords.words('english'))  # Set of English stopwords
```

```python
for column in columns_to_analyze:
    print(f"Analysis for Column: {column}\n")
    word_counts_per_row = Counter()  # Counter to store word counts per row

    for index, row in df.iterrows():
        text = row[column]
        words = re.findall(r'\b[a-zA-Z]+\b', str(text))
        unique_words = set([word.lower() for word in words if word.lower() not in stop_words])
        
        word_counts_per_row.update(unique_words)

    # Get the top 5 words mentioned the most across rows for this column
    top_words = word_counts_per_row.most_common(5)

    # Print the top 5 words
    print("Top 5 words mentioned the most:")
    for word, count in top_words:
        print(f"Word: {word}, Total Mentions: {count}")

    # Print rows that include the top 5 words for this column
    for word, _ in top_words:
        filtered_rows = df[df[column].str.lower().str.contains(word)]
        print(f"\nRows mentioning '{word}' in Column {column}:")
        for index, row in filtered_rows.iterrows():
            row_number = index + 1
            response_value = row[column]
            print(f"Row {row_number}: {response_value}")
        print("\n")  # Separate output for each word
```

```python
# Create a dictionary with Q8
data_q10 = {
    'Word': ['course', 'would', 'great', 'people', 'group'],
    'Total Mentions': [10, 6, 5, 4, 4]
}

# Create a DataFrame from the dictionary
df_q10 = pd.DataFrame(data_q10)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create a function to generate and display word clouds
def generate_word_cloud(data, title):
    # Combine words and their total mentions into a dictionary
    word_mentions = {row['Word']: row['Total Mentions'] for _, row in data.iterrows()}

    # Create a WordCloud object
    wordcloud = WordCloud(width=800, height=400, background_color='white', prefer_horizontal=1.0, min_font_size=10)

    # Generate the word cloud using a dictionary of word frequencies and font sizes
    wordcloud.generate_from_frequencies(word_mentions)

    # Display the word cloud with a title using Matplotlib
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(title)
    plt.axis("off")
    plt.show()

# Generate word clouds for each DataFrame
#generate_word_cloud(df_q8, "Word Cloud for Q8")
#generate_word_cloud(df_q9, "Word Cloud for Q9")
generate_word_cloud(df_q10, "Word Cloud for Q10")
#generate_word_cloud(df_q11, "Word Cloud for Q11")
```
